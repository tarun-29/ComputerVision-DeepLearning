```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
#lets draw rectangle on image
cv2.rectangle(blank_img,pt1=(384,0),pt2=(510,150), color=(0,255,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img,center=(100,100),radius=50, color=(255,0,0),thickness=8)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img,center=(400,400),radius=50, color=(255,0,0),thickness=-1)
# thickness negative means to fill circle rectangle
plt.imshow(blank_img)
```

```python
#draw a line
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text="Tarun",org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(blank_img)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
```

```python
vertices.shape
#open cv use 3 three dimesnions so cast this array of (4,2) into 3 dimension
```

```python
pts = vertices.reshape((-1,1,2))
vertices.shape
pts.shape
pts
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(blank_img)
```
