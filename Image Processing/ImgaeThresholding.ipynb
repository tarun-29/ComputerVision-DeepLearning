```python
#threshold convert image in white and black
# imgae -> grayscale ->threshold image (binary)
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../DATA/rainbow.jpg",0)
plt.imshow(img)
```

```python
ret,thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
# ret,thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
ret
```

```python
plt.imshow(thresh1,cmap="gray")
```

```python
img = cv2.imread("../DATA/crossword.jpg",0)
plt.imshow(img,cmap="gray")
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap="gray")
```

```python
show_pic(img)
```

```python
# A simple binary threshold
ret,th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img, 255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```python
blended = cv2.addWeighted(src1=th1, alpha=0.6, src2=th2, beta=0.4,gamma=0)
show_pic(blended)
```
