```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../DATA/00-puppy.jpg")
plt.imshow(img) 
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#convert this image into hsv hsl
img = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
plt.imshow(img)
```
