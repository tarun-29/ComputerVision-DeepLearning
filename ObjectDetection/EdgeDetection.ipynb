```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../DATA/sammy_face.jpg")
```

```python
plt.imshow(img)
```

```python
edge = cv2.Canny(image=img, threshold1=127, threshold2=127)
```

```python
plt.imshow(edge)
```

```python
med_value = np.median(img)
```

```python
med_value
```

```python
#LOWER THRESHOLD TO EITHER 0 OR 70% OF THE MEDIAN VALUE WHICHEVER IS GREATER
lower = int(max(0,0.7*med_value))
#UPPER THRESHOLD TO EITHER 130% OF THE MEDIAN OR MAX 255 WHICHEVER IS SMALLER
upper = int(min(255,1.3*med_value))
```

```python
edge = cv2.Canny(image=img, threshold1=lower, threshold2=upper+100)
plt.imshow(edge)
```

```python
blurred_image = cv2.blur(img, ksize=(8,8))
```

```python
edge = cv2.Canny(image=blurred_image, threshold1=lower, threshold2=upper)
plt.imshow(edge)
```
